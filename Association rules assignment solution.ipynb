```python
from mlxtend.preprocessing import TransactionEncoder
transactions=[]
for i in range(len(raw_data)):
    # Read the first cell of each row (Index 0)
    row_string = str(raw_data.iloc[i, 0]) 
    # Split by comma and clean up whitespace/empty values
    items = [item.strip() for item in row_string.split(',') if item.strip() and item.strip().lower() != 'nan']
    if items:
        transactions.append(items)
#print(transactions)
te=TransactionEncoder()
te_array=te.fit(transactions).transform(transactions)
processed_data=pd.DataFrame(te_array,columns=te.columns_)
processed_data.head(10)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

freq_itemsets = apriori(processed_data,min_support=0.003,use_colnames=True)
#print(freq_itemsets.head())

rules=association_rules(freq_itemsets,metric="lift",min_threshold=3)
#print(rules.head())


filtered_rule=rules[rules['confidence'] >= 0.02]

#print(filtered_rule)
print(filtered_rule[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values('lift', ascending=False).head(10))
```

```python
# Support (0.003): This requires an item combination to appear at least 22 times in the dataset. Setting this too high (e.g., 0.05) would only return very common items like "mineral water," missing the "hidden" associations.

# Confidence (0.2): This means that in 20% of the cases where the first item is bought, the second is also bought. This is a standard starting point for retail data where basket variety is high.

# Lift (> 3): This is the most critical threshold. A lift of 3 means the items are bought together 3 times more often than would be expected by random chance. Any lift > 1 indicates a positive relationship.
```
