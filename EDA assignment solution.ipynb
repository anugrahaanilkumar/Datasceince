```python
import matplotlib.pyplot as plt
n=len(data.columns)
for i in range(0,n):
    plt.figure(figsize=(8,8))
    data.boxplot(column=[data.columns[i]])
    plt.ylabel("Count")
    plt.show()
```

Analysing the boxplots we could confirm the presence of outliers in most of the columns, so we need to remove the outliers

***Replacing NULL values with median***

```python
data.isnull().sum()
```

```python
for i in data.columns:
    median_col=data[i].median()
    data[i].replace(np.nan,median_col,inplace=True)
    print(median_col)
    print(data[i])

data.isnull().sum()
```

***Removing outliers***

```python
data.shape
```

```python
for col in data.columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    data_cleaned = data[(data[col]>=lower_range) & (data[col]<=upper_range)]

print(data.shape)
print(data_cleaned.shape)
```
