```python
import scipy.stats as stats
import numpy as np

pop_mean=4000
sample_mean=3050
sample_size=25
std_dev=5*sample_size
test_statistic=(sample_mean-pop_mean)/(std_dev/np.sqrt(sample_size))
print(test_statistic)
```

***3. Determine the Critical Value
Using the alpha level of 5% (α = 0.05), determine the critical value from the standard normal (Z) distribution table.***

```python
#For a right-tailed test where alpha = 0.05, we look for the Z-score where the area to the left ie, 1-alpha=1-0.05=0.95
alpha=0.05
z_critical=stats.norm.ppf(1-alpha)
print(z_critical)
```

***4. Make a Decision:
Compare the test statistic with the critical value to decide whether to reject the null hypothesis.***

```python
if test_statistic>z_critical:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```
