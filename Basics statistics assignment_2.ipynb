***Build 99% Confidence Interval Using Sample Standard Deviation. Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.***

```python
import numpy as np
import scipy.stats as stats
```

```python
#sample data
data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

Formula for t-distribution confidence interval = sample mean Â± t_score * standard error of mean

standard error of mean(sem)=standard deviation/square root of n

```python
#calculating standard deviation of sample
std_dev=np.std(data)
#calculating mean of sample
mean=np.mean(data)
#calculating length of sample
n=len(data)
#calculating degree of freedom of sample
df=n-1
#calculating alpha ie, alpha(significance level)=1-confidence level
confidence_level=0.99
alpha=1-confidence_level
```

```python
#calculating t-score or critical value
t_score=stats.t.ppf(1-(alpha/2),df)
#calculating standard error of mean
sem=std_dev/np.sqrt(n)
#calculating margin of error
margin_of_error=t_score*sem
#calculating confidence interval values
upper_value=mean+margin_of_error
lower_value=mean-margin_of_error
print(f"Standard devaiation:{std_dev}")
print(f"t-score:{t_score}")
print(f"Margin of error:{margin_of_error}")
print(f"Confidence interval:({lower_value},{upper_value})")
```

Therfore, the the mean number of characters printed before the print-head fails will be within the interval (1.095231668626123,1.3821016647072102)

We used t-distribution for this task since the size of the sample is less than 30 (here size is 15) and standard deviation is also unknown and thus t-distribution is used in such cases.
