```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=df.drop(['letter'],axis=1)
Y=df['letter']
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test=train_test_split(X_scaled,Y,test_size=.03,random_state=42)
print(X_train)
print(X_train.shape)
print(Y_train)
print(Y_train.shape)
print(X_test)
print(X_test.shape)
print(Y_test)
print(Y_test.shape)
```

**2. Model Implementation**

**●	Construct a basic ANN model using your chosen high-level neural network library. Ensure your model includes at least one hidden layer.**

**●	Divide the dataset into training and test sets.**

**●	Train your model on the training set and then use it to make predictions on the test set.**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Constructing the Basic ANN Model
model = Sequential([
    # Input Layer + First Hidden Layer
    # We use 'relu' for hidden layers as it's efficient and prevents vanishing gradients
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)), 
    
    # Second Hidden Layer (As requested: "at least one hidden layer")
    Dense(32, activation='relu'),
    
    # Output Layer
    # Must have 26 neurons (for A-Z) and use 'softmax' for probability distribution
    Dense(26, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

# Train the model
# Epochs: number of times the model sees the full dataset
# Batch_size: number of samples processed before updating weights
history = model.fit(X_train, Y_train, epochs=20, batch_size=32, validation_split=0.1, verbose=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 1. Generate probability predictions for the test set
# This returns a 2D array: [number_of_samples, 26]
y_probs = model.predict(X_test)

# 2. Convert probabilities to class integers (0-25) 
# np.argmax picks the index with the highest probability
y_pred = np.argmax(y_probs, axis=1)

# 3. Print the Final Accuracy
test_acc = accuracy_score(Y_test, y_pred)
print(f"Final Test Accuracy: {test_acc:.2%}")

# 4. Detailed Classification Report
# We use le.classes_ to show the actual letters ('A', 'B'...) instead of 0, 1...
print("\nClassification Report:")
print(classification_report(Y_test, y_pred, target_names=le.classes_))
```

**●	Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.**

```python
from tensorflow.keras.optimizers import Adam

# Define different hyperparameter sets to test
experiments = {
    "Baseline": {"layers": [64, 32], "lr": 0.001, "act": "relu"},
    "Deeper": {"layers": [128, 64, 32], "lr": 0.001, "act": "relu"},
    "Slow_Learner": {"layers": [64, 32], "lr": 0.0001, "act": "relu"},
    "Tanh_Activation": {"layers": [64, 32], "lr": 0.001, "act": "tanh"}
}

results = {}

for name, params in experiments.items():
    print(f"\n--- Running Experiment: {name} ---")
    
    # Construct model dynamically based on params
    model1 = Sequential()
    # Input Layer
    model1.add(Dense(params["layers"][0], activation=params["act"], input_shape=(X_train.shape[1],)))
    
    # Hidden Layers
    for units in params["layers"][1:]:
        model1.add(Dense(units, activation=params["act"]))
        
    # Output Layer
    model1.add(Dense(26, activation='softmax'))
    
    # Compile with specific learning rate
    model1.compile(optimizer=Adam(learning_rate=params["lr"]), 
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])
    
    # Train
    history = model1.fit(X_train, Y_train, epochs=15, batch_size=32, validation_split=0.1, verbose=0)
    
    # Store Final Validation Accuracy
    final_val_acc = history.history['val_accuracy'][-1]
    results[name] = final_val_acc
    print(f"Final Validation Accuracy: {final_val_acc:.4f}")

# Display Comparison
print("\n--- Final Hyperparameter Comparison ---")
for name, acc in results.items():
    print(f"{name}: {acc:.2%}")
```

**●	Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.**

```python

# 1. Define the parameter grid
param_grid = {
    'neurons': [64, 128],
    'activation': ['relu', 'tanh'],
    'learning_rate': [0.01, 0.001] # Keeping it small for speed, add more if needed
}

grid_results = []

# 2. Structured Loop (Manual Grid Search)
for n in param_grid['neurons']:
    for a in param_grid['activation']:
        for lr in param_grid['learning_rate']:
            
            print(f"Testing: Neurons={n}, Act={a}, LR={lr}")
            
            # Build model
            model2 = Sequential([
                Dense(n, activation=a, input_shape=(X_train.shape[1],)),
                Dense(32, activation=a),
                Dense(26, activation='softmax')
            ])
            
            model2.compile(optimizer=Adam(learning_rate=lr),
                          loss='sparse_categorical_crossentropy',
                          metrics=['accuracy'])
            
            # Train (fewer epochs for searching phase)
            history1 = model2.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=0, validation_split=0.1)
            
            # Record results
            val_acc = history1.history['val_accuracy'][-1]
            grid_results.append({
                'neurons': n,
                'activation': a,
                'learning_rate': lr,
                'val_accuracy': val_acc
            })

# 3. Convert results to DataFrame for documentation
df_results = pd.DataFrame(grid_results).sort_values(by='val_accuracy', ascending=False)
print("\n--- Grid Search Results ---")
print(df_results)
```

**●	Employ suitable metrics such as accuracy, precision, recall, and F1-score to evaluate your model's performance.**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# 1. Get predictions
y_probs = model.predict(X_test)
y_pred = np.argmax(y_probs, axis=1)

# 2. Calculate global metrics (weighted to account for any class imbalance)
accuracy = accuracy_score(Y_test, y_pred)
precision = precision_score(Y_test, y_pred, average='weighted')
recall = recall_score(Y_test, y_pred, average='weighted')
f1 = f1_score(Y_test, y_pred, average='weighted')

print(f"--- Global Model Performance ---")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")

# 3. Class-wise breakdown (The detailed report)
print("\n--- Detailed Classification Report ---")
print(classification_report(Y_test, y_pred, target_names=le.classes_))
```

### **2. Why the "Deeper" Model performed better**

#### **A. Increased Feature Resolution (The "128" Factor)**
By doubling the neurons in the first layer from **64 to 128**,  increased the model's **width**.
* **Effect:** The model can now track 128 different "filters" or shapes at once. For the alphabet dataset, this means it can simultaneously look for more subtle features—like the exact angle of the slant in a 'K' vs. an 'X'—that the 64-neuron model might have averaged together.

#### **B. Hierarchical Depth (The 3rd Layer)**
Adding a third hidden layer creates a deeper **hierarchy of abstraction**.
* **Initial Model:** Detected edges, then combined them into basic shapes.
* **Deeper Model:** Detects edges, combines them into shapes, and then uses the third layer to analyze the **spatial relationship** between those shapes. This is likely how it squeezed out that extra 0.33%, by better distinguishing between letters that share almost all the same parts (like 'E', 'F', and 'L').


### **3. The Diminishing Returns of Tuning**
The fact that you only gained **0.33%** despite nearly tripling the parameter count illustrates a key concept in Deep Learning: **The Law of Diminishing Returns.**

* **Observation:** Your Initial Model was already "saturated"—it had enough power to solve most of the dataset. 
* **Tuning Role:** The Deeper Model's "significant role" wasn't to learn the alphabets from scratch, but to resolve the **residual errors**. It likely fixed 2 or 3 specific samples (like a very messy 'H' that looked like an 'M') that the simpler model simply didn't have the mathematical "memory" to store.

### **4. Analysis of the Specific "Deeper" Config**
1.  **Tapered Architecture:** By reducing neurons in each layer (128 → 64 → 32),it created a "funnel" that forces the model to compress the most important information, preventing it from just memorizing the data.
2.  **Stable Learning Rate:** Keeping the learning rate at `0.001` with a deeper network is safer than increasing it, as deeper networks are more prone to "Exploding Gradients" if the steps are too large.
